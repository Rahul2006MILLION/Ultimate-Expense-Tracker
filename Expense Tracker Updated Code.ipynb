```python
all_amt=[]
rec_food=[]
rec_food_amt=[]
rec_shop=[]
rec_shop_amt=[]
rec_travel=[]
rec_travel_amt=[]
rec_other=[]
rec_other_amt=[]
choice=''
while True:
    try:
        choice=int(input("1.Add Expense \n2.View All Expenses \n3.Search by Category \n4.Clear the List \n5.Exit: "))
    except:
        print("INVALID INPUT!!")
    else:
        
        if choice==1:
            cat=input("Enter Expence Type: (Food/Travel/Shopping/Others): ").lower()
            if cat=='food':
                with open("q.txt",'a') as f:
                    try:
                        w=int(input("Enter Number of Food Expences: "))
                    except:
                        print("Invalid Input, Please Enter Number of Food Expences")
    
                    else:
                        for i in range(w):
                            try:
                                amt=int(input('Enter Amount $: '))
                                note=input("Enter Note: ")
                            except:
                                print("Please Input Amount(integer) then Note")
                            else:
                                f.write(f"Note: {note} | ${amt}/-\n ")
                                all_amt.append(amt)
                                rec_food.append(f"Note: {note} | ${amt}/-\n ")
                                rec_food_amt.append(amt)
                                    
            elif cat=="travel":
                with open("q.txt",'a') as d:
                    try:
                        e=int(input("Enter the Number of Travel Expences: "))
                    except:
                        print("Invalid Input, Please Enter Number of Travel Expences")
    
                    else:
                        for j in range(e):
                            try:
                                amtt=int(input("Enter Amount $: "))
                                notee=input("Enter Note: ")
                            except:
                                print("Please Input Amount(integer) then Note")
                            else:
                                d.write(f"Note: {notee} | ${amtt}/-\n")
                                all_amt.append(amtt)
                                rec_travel.append(f"Note: {notee} | ${amtt}/-\n")
                                rec_travel_amt.append(amtt)
            
            elif cat=="shopping":
                with open("q.txt",'a') as tt:
                    try:
                        r=int(input("Enter the Number of Shopping Expences: "))
                    except:
                        print("Invalid Input, Please Enter Number of Shopping Expences")
                    else:
                        for k in range(r):
                            try:
                                ammt=int(input("Enter Amount $: "))
                                nottee=input("Enter Note: ")
                            except:
                                print("Please Input Amount(integer) then Note")
                            else:
                                tt.write(f"Note: {nottee} | ${ammt}/-\n")
                                all_amt.append(ammt)
                                rec_shop.append(f"Note: {nottee} | ${ammt}/-\n")
                                rec_shop_amt.append(ammt)
                        
            elif cat=="others": 
                with open("q.txt",'a') as h:
                    try:
                        p=int(input("Enter the Number of Other Expences: "))
                    except:
                        print("Invalid Input, Please Enter Number of Other Expences")
                    else:
                        for m in range(p):
                            try:
                                aamt=int(input("Enter Amount $: "))
                                noote=input("Enter Note: ")
                            except:
                                print("Please Input Amount(integer) then Note")
                            else:
                                h.write(f"Note: {noote} | ${aamt}/-\n")
                                all_amt.append(aamt)
                                rec_other.append(f"Note: {noote} | ${aamt}/-\n")
                                rec_other_amt.append(aamt)

            else:
                print('\n---Invalid Expence Type--\n')

          
                        
        elif choice==2:
            with open("q.txt",'r') as c:
                contents=c.readlines()
                if not contents:
                    print("\n----No Expences Added----\n")
                else:
                    for num,con in enumerate(contents,1):
                        print(f"{num}. {con}")
                    print('---'*10)
                    print(f"Total Expenditure: ${sum(all_amt)}")
    
        elif choice==3:
            category=int(input("Enter Particular Expenditure\n1.Food\n2.Travel\n3.Expenditure\n4.Others:\nEnter (1/2/3/4): "))
            if category==1:
                if not rec_food:
                    print("\n---No Food Expences Added Here---\n")
                else:
                    for j in range(len(rec_food)):
                        print(rec_food[j])
                    print('--'*10)
                    print(f"Total Expenditure on Food: ${sum(rec_food_amt)}")
    
            elif category==2:
                if not rec_travel:
                    print("\n---No Travel Expences Added Here---\n")
                else:
                    for g in range(len(rec_travel)):
                        print(rec_travel[g])
                    print('--'*10)
                    print(f"Total Expenditure on Travel: ${sum(rec_travel_amt)}\n")
    
            elif category==3:
                if not rec_shop:
                    print("\n---No Shopping Expences Added Here---\n")
                else:
                    for jj in range(len(rec_shop)):
                        print(rec_shop[jj])
                    print('--'*10)
                    print(f"Total Expenditure on Shopping: ${sum(rec_shop_amt)}")
    
            elif category==4:
                if not rec_other:
                    print("\n---No Other Expences Added Here---\n")
                else:
                    for t in range(len(rec_other)):
                        print(rec_other[t])
                    print('--'*10)
                    print(f"Total Expenditure on Other Items: ${sum(rec_other_amt)}")
    
            else:
                print("Invalid Input :(")
    
        elif choice==4:
            with open("q.txt",'w') as l:
                all_amt.clear()
                rec_food.clear()
                rec_food_amt.clear()
                rec_shop.clear()
                rec_shop_amt.clear()
                rec_travel.clear()
                rec_travel_amt.clear()
                rec_other.clear()
                rec_other_amt.clear()
                print("\n---MEMORY CLEARED---\n")
        
        elif choice==5:
            print("\n---EXPENCE TRACKER EXITING---\n")
            break
    
        else:
            print("\n---Invalid Input:[---\n")
```
